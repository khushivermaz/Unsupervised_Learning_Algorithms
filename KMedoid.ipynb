```python
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Generate sample data
X, y = make_blobs(n_samples=500, centers=3, random_state=0)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1], c='blue', marker='o', s=30)
plt.grid()
plt.show()
```

```python
# Cluster the data using KMedoids
kmedoids = KMedoids(n_clusters=3, random_state=0)
kmedoids.fit(X)
```

```python
print(kmedoids.labels_.shape)
```

```python
# Plot the results
plt.figure(figsize=(7.5, 3.5))
plt.scatter(X[:, 0], X[:, 1], c=kmedoids.labels_, cmap='viridis')
plt.scatter(kmedoids.cluster_centers_[:, 0],kmedoids.cluster_centers_[:, 1], marker='x', color='red')
plt.show()
```
