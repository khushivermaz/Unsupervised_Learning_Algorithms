```python
#Davis Bouldin Index
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt
```

```python
# load the datasets
iris = load_iris()
data = iris.data
```

```python
# perfrom kmeans clustring
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit(data)
```

```python
# get the cluster labels
labels = kmeans.labels_
```

```python
# calculate DBI
dbi = davies_bouldin_score(data, labels)
print("Davies-Bouldin Index:", dbi)
```

```python
#Extract centroids of each cluster
centroids = kmeans.cluster_centers_
```

```python
#plot the clustering results in a scatter plot
plt.figure(figsize=(8, 6))
```

```python
#Scatter plt the data points with different colors for each cluster

for i in range(3):
  cluster_data = data[labels == i]
  plt.scatter(cluster_data[:,0], cluster_data[:,1], label = f'Cluster { i + 1}')

#plot the centroids as well
plt.scatter(centroids[:,0], centroids[:,1], s =200, c = 'k', marker = 'X', label = 'Centroids')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title('K-Means Clustering Results')
plt.legend()
plt.show()
```
