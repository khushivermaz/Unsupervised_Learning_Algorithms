```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
X,y = make_blobs(n_samples= 300, n_features= 5, centers= 3, random_state= 42)
```

```python
print("Original shape of the dataset: " ,X.shape)
```

```python
#perform pca to reduce dimensions to 2 for visualization
pca = PCA(n_components= 2)
X_reduced = pca.fit_transform(X)
```

```python
print(X_reduced[:,0])
```

```python
print(X_reduced[:,1])
```

```python
# display the shape of dataset after PCA
print("Reduced Shape of the dataset after PCA: " ,X_reduced.shape)
```

```python
#plot the reduced dataset
plt.figure(figsize= (8,6))
plt.scatter(X_reduced[:,0], X_reduced[:,1], c= y, cmap='viridis', edgecolor= "k")
plt.xlabel("Principle component 1")
plt.ylabel("Principle component 2")
plt.title("PCA on make_blobs Dataset")
plt.colorbar(label= "Cluster Label")
plt.show()
```
